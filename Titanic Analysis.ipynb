# Titanic Analysis

## Introduction

The Titanic dataset contains information on 891 passengers aboard the RMS Titanic, including demographics, travel details, and survival status. It is commonly used to explore which factors—such as age, gender, and passenger class—affected the likelihood of survival. This dataset provides a simple yet powerful foundation for learning data analysis and building predictive models.

## Data Description
* PassengerID: Unique identifier for each passenger.
* Pclass: Passenger class (1st, 2nd, 3rd).
* Name: Full name of the passenger.
* Sex: Gender of the passenger.
* Age: Age of the passenger.
* SibSp: Number of siblings/spouses aboard.
* nParch: Number of parents/children aboard.
* Ticket: Ticket number.
* Fare: Ticket fare.
* Cabin: Cabin number.
* Embarked: Port of embarkation (C = Cherbourg; Q = Queenstown; S = Southampton).
* Survived: Survival status (0 = No; 1 = Yes).

### Fill Null Values

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Cabin'] = df['Cabin'].fillna('Unknown')
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

## Pandas Filtering Questions

#### 1) Get the details of passengers who survived and were younger than 18 years old.

```python
df[(df['Survived']==1) & (df['Age']<18)]
```

#### 2) Get the details of male passengers who were older than 40 years old

```python
df[(df['Sex']=='male') & (df['Age']>40)]
```

#### 3) Get the details of passengers who paid a fare greater than $30 and survived.

```python
df[(df['Fare']>30) & (df['Survived']==1)]
```

#### 4) Get the details of female passengers who did not survive.

```python
df[(df['Sex']=='female') & (df['Survived']==0)]
```

#### 5) Get the details of passengers who were in either the first or second class.

```python
df[(df['Pclass']==1) | (df['Pclass']==2)]
```

#### 6) Get the details of passengers who had more than two siblings/spouses aboard the        Titanic and did not survive.

```python
df[(df['SibSp']>2) & (df['Survived']==0)]
```

#### 7) Get the details of passengers who were in the third class and were older than 30 years old.

```python
df[(df['Pclass']==3) & (df['Age']>30)]
```

#### 8) Get the details of passengers who were younger than 15 years old and had parents/children aboard.

```python
df[(df['Age']<15) & (df['Parch']>0)]
```

#### 9) Get the details of male passengers who paid less than $10 for their fare.

```python
df[(df['Sex']=='male') & (df['Fare']<10)]
```

#### 10) Get the details of passengers who were in the third class, younger than 20, and had siblings/spouses aboard.

```python
df[(df['Pclass']==3) & (df['Age']<20) & (df['SibSp']>0)]
```

```python
df[((df['Age']<10) | (df['Age']>60)) & (df['Survived']==1)]
```
